# Experiment 1: XOR Gate using Artificial Neural Network (ANN)

## Objective
To implement an Artificial Neural Network using TensorFlow/Keras to learn the XOR logic gate.

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y = np.array([
    [0],
    [1],
    [1],
    [0]
])

model = Sequential()
model.add(Dense(4, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(
    loss='binary_crossentropy',
    optimizer=Adam(learning_rate=0.1),
    metrics=['accuracy']
)
model.fit(X, y, epochs=1000, verbose=0)
predictions = model.predict(X)
print("Predictions for XOR inputs:")
for i in range(len(X)):
    print(
        "Input:", X[i],
        "Predicted output:", predictions[i][0],
        "Actual output:", y[i][0]
    )
loss, accuracy = model.evaluate(X, y, verbose=0)
print("Model accuracy:", accuracy * 100, "%")
```
